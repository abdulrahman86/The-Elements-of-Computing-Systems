```python
for i in range(1, 111):
    fizz = 3
    buzz = 5
    
    result = ""
    
    if i%fizz == 0:
        result += "fizz"
    if i%buzz == 0:
        result += "buzz"
    if len(result) == 0:
        result = str(i)
    print(result)
```
